### Analizar las ventas mensuales

```python
data_trabajo.head()
```

```python
data_trabajo['Order Date'][3].split(' ')
```

```python
data_trabajo['Order Date'][0].split(' ')[0].split('/')[0] # mes
```

```python
data_trabajo['Order Date'][3].split(' ')[0].split('/')[2] # año
```

```python
def month(a):
    return a.split(' ')[0].split('/')[0]
```

```python
def year(a):
    return a.split(' ')[0].split('/')[2]
```

```python
import warnings
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
data_trabajo = data_trabajo[data_trabajo['Order Date']!='Order Date']
```

```python
data_trabajo['month']= data_trabajo['Order Date'].apply(month)
data_trabajo['month'] = data_trabajo['month'].astype(int)
```

```python
data_trabajo['year']=data_trabajo['Order Date'].apply(year)
```

```python
data_trabajo.replace(['19','20'], ['2019','2020'], inplace=True)
```

```python
data_trabajo.head()
```

```python
data_trabajo['sales'] = data_trabajo['Quantity Ordered'].astype(int) * data_trabajo['Price Each'].astype(float)
```

```python
data_trabajo1 = pd.DataFrame({'month': data_trabajo['month'],'year': data_trabajo['year'], 'sales': data_trabajo['sales']})
```

```python
data_trabajo1.head()
```

```python
data_trabajo1 = data_trabajo1.groupby(['year','month']).sum()
data_trabajo1 = data_trabajo1.reset_index()
data_trabajo1
```

La cantidad de ventas realizadas en 2020 es muy pequeña, por lo tanto, podemos despreciarlo y solo tomar las ventas realizadas en el año de 2019.

```python
data_trabajo1 = data_trabajo1[:12]
```

```python
data_trabajo1.info()
```

```python
sns.catplot(data = data_trabajo1, x= 'month', y= 'sales', kind= 'bar', hue='year')
plt.show()
```

* Se observa que hay mayor cantidad de ventas en el mes de diciembre. Esto puede darse debido a la compra de regalos para el día de Navidad.
* La mayor cantidad de ventas puede darse debido a los descuentos de los diferentes objetos por Navidad y año nuevo.

### ¿Cuál ciudad tiene más pedidos?

```python
data_trabajo_sin_2020 = data_trabajo[data_trabajo['year'] != '2020']
```

```python
data_trabajo_sin_2020['Purchase Address'][0].split(',')
```

```python
data_trabajo_sin_2020['Purchase Address'][0].split(',')[1]
```

```python
def city(a):
    return a.split(',')[1]
```

```python
data_trabajo_sin_2020['city'] = data_trabajo_sin_2020['Purchase Address'].apply(city)
```

```python
data_trabajo_sin_2020.head()
```

```python
data_trabajo2 = pd.DataFrame({'city': data_trabajo_sin_2020['city'], 'Order ID': data_trabajo_sin_2020['Order ID']})
```

```python
data_trabajo2 = data_trabajo2.groupby('city')['Order ID'].count()
```

```python
data_trabajo2 = data_trabajo2.reset_index()
data_trabajo2 = data_trabajo2.rename(columns={"Order ID": "Quantity Customers"})
```

```python
data_trabajo2 
```

```python
plt.pie(data= data_trabajo2, x = 'Quantity Customers', labels= 'city', autopct='%1.0f%%')
plt.show()
```

* La ciudad que tiene más pedidos es San Francisco.

### ¿Qué producto es más vendido? ¿Por qué?

```python
data_trabajo_sin_2020
```

```python
data_trabajo3 = pd.DataFrame({'Product': data_trabajo_sin_2020['Product'], 'Quantity Ordered': data_trabajo_sin_2020['Quantity Ordered'], 'Price Each': data_trabajo_sin_2020['Price Each']})
data_trabajo3.head()
```

```python
data_trabajo3['Quantity Ordered'] = data_trabajo3['Quantity Ordered'].astype(int)
data_trabajo3['Price Each'] = data_trabajo3['Price Each'].astype(float)
```

```python
data_trabajo3 = data_trabajo3.groupby('Product').agg({'Quantity Ordered': 'sum', 'Price Each': 'mean'})
data_trabajo3 = data_trabajo3.reset_index()
```

```python
data_trabajo3 
```

```python
data_trabajo3.sort_values('Quantity Ordered', ascending= False)
```

```python
data_trabajo3[['Quantity Ordered', 'Price Each']].corr()
```

*  Se observa que hay una correlación negativa entre la cantidad ordenada y el precio de cada producto. Por lo tanto, uno de ellos sube y el otro baja.

```python
fig , ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(data_trabajo3['Product'],data_trabajo3['Quantity Ordered'], color = 'red')
ax2.plot(data_trabajo3['Product'],data_trabajo3['Price Each'])
ax1.set_xticklabels(data_trabajo3['Product'], rotation = 'vertical', fontsize = 16)
ax1.set_ylabel('Cantidad ordenada')
ax2.set_ylabel('Media del precio de los productos')

plt.show()
```

* El producto más vendido es AAA Batteries (4-pack).
* Se observa que para los objeto de menor precio son los más vendidos que los productos con mayor precio.

### ¿Qué productos suelen venderse juntos?

```python
data_trabajo_sin_2020.drop('grouped_products', axis=1,inplace= True)
```

```python
data_trabajo_sin_2020['Order ID'].duplicated(keep= False)
```

```python
duplicadosID = data_trabajo_sin_2020[data_trabajo_sin_2020['Order ID'].duplicated(keep= False)]
duplicadosID.head()
```

```python
agrupar_productos = duplicadosID.groupby(['Order ID'])['Product'].apply(lambda x: ','.join(x)).reset_index().rename(columns= {'Product' :  'grouped_products'})
agrupar_productos
```

```python
agrupando_productos = duplicadosID.merge(agrupar_productos,  how= 'left', on= 'Order ID')
agrupando_productos
```

```python
elimnar_duplicados = agrupando_productos.drop_duplicates(subset= ['Order ID'])
elimnar_duplicados
```

```python
data_trabajo6 = elimnar_duplicados.groupby(['grouped_products'])[['grouped_products']].value_counts().reset_index()
data_trabajo6 = data_trabajo6.sort_values('count', ascending= False)
data_trabajo6 = data_trabajo6.iloc[0:5]
data_trabajo6
```

```python
plt.pie(data= data_trabajo6, x = 'count', labels= 'grouped_products', autopct='%1.0f%%')
plt.show()
```

* Se observa que el producto que sule venderse juntos mayormente es iPhone,Lightning Charging Cable.
